In this quick project we are performing an insight analysis on the Weather data, by answering the 10 questions

```
import pandas as pd

w_data = pd.read_csv("/work/cleaned_weather.csv")

w_data.head(2)
```

```
w_data.rename(columns = {'DATE':'n_date','PRCP':'inches_rain','TMAX':'temp_max','TMIN':'temp_min'}, inplace = True)
w_data['n_date']= pd.to_datetime(w_data['n_date'])
w_data.head(2)
```

1. Select all rows from December 1st, 2000 to December 15th, 2000

```


df_1 = _deepnote_execute_sql("""select *
from w_data
where n_date between '2000-12-01' and '2000-12-15'
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_1
```

2. Get the average maximum temperature for every year from the year 2000 onward. Order the results by year

```


df_2 = _deepnote_execute_sql("""with tt as (select year(n_date) as year,
round(avg(temp_max),2) as avg_max_temp
from w_data
group by 1)

select year,
avg_max_temp
from tt
where year > 2000
order by 1
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_2
```

3. Get the standard deviation of the maximum temperature per year, from 2000 onward. Order by year

```


df_3 = _deepnote_execute_sql("""with tt as (select year(n_date) as year,
round(stddev(temp_max),2) as sd_max_temp
from w_data
group by 1)

select year,
sd_max_temp
from tt
where year > 2000
order by 1
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_3
```

4. What are the 10 hottest days on record? Take hottest to mean 'highest maximum temperature

```


df_4 = _deepnote_execute_sql("""select n_date,
avg(temp_max)
from w_data
group by 1
order by 2 desc
limit 10
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_4
```

5. In 2016, what fraction of days did it rain?

```


df_5 = _deepnote_execute_sql("""with tt as (select year(n_date) as year,
rain
from w_data
where year = 2016)

select
100*count(*)/(select count(*) from tt) as Rain_prcnt
from tt
where year = 2016 and rain = 'True'
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_5
```

6. Get the 10 years with the hottest average maximum temperature in July. Order from hottest to 
coolest

```


df_6 = _deepnote_execute_sql("""select year, avg(temp_max)
from (select year(n_date) as year, month(n_date) as month, temp_max
from w_data) as sbq
where month = 7
group by 1
order by 2 desc
limit 10
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_6
```

7. Get the 10 years with the coldest average minimum temperature in December. Order from 
coolest to hottest

```


df_7 = _deepnote_execute_sql("""select year, avg(temp_min)
from (select year(n_date) as year, month(n_date) as month, temp_min
from w_data) as sbq
where month = 12
group by 1
order by 2 asc
limit 10
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_7
```

 8. Repeat the last question, but round the temperatures to 3 decimal places

```


df_8 = _deepnote_execute_sql("""select year, round(avg(temp_min),3)
from (select year(n_date) as year, month(n_date) as month, temp_min
from w_data) as sbq
where month = 12
group by 1
order by 2 asc
limit 10
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_8
```

10. Give the average inches of rain that fell per day for each month, where the average is taken over 
2000 - 2010 (inclusive)

```


df_9 = _deepnote_execute_sql("""select year,month, rain_avg
from (select year(n_date) as year,month(n_date) as month, round(avg(inches_rain),3) as rain_avg
from w_data
where rain='True'
group by 1,2) as sbq
where year>=2000 and year<=2010
""", 'SQL_DEEPNOTE_DATAFRAME_SQL')
df_9
```
